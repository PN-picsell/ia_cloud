```python
import random  as rand
from numpy.random import choice
import collections

def choose(array):
    y = random()
    lastLetter = array[-1]
    if lastLetter == "e":
        if y<0.1:
            array+="e"
        elif y>0.8:
            array+="a"
        else:
            array+="t"
    elif lastLetter == "a":
        if y<0.1:
            array+="a"
        elif y>0.6:
            array+="t"
        else:
            array+="e"
    else:
        if y<0.1:
            array+="t"
        elif y>0.8:
            array+="e"
        else:
            array+="a"

```
