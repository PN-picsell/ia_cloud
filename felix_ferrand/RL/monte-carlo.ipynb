```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
hists = []
for i in range(10000):
    hist = []
    hist.append(np.random.gamma(2.0, 2.0, 1000))
    x, b = np.histogram(hist, bins=100)
    hists.append(x)

result = np.mean(hists, axis = 0)

sn.lineplot(np.linspace(0, 1, len(result)), hists)
```
