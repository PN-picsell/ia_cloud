```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.io import imsave
```

```python
def convol(img, kernel):
    x_size=len(img[0])
    y_size=len(img)

    img_conv = np.zeros((y_size, x_size))

    # convolution
    for x in range(1, x_size):
        for y in range(1, y_size):
            pixel_gray = 0
            for i in range(0,len(kernel),1):
                for j in range(0,len(kernel[0]),1):
                    pixel_gray = pixel_gray + img[y-j][x-i]*kernel[i][j]
            img_conv[y][x] = pixel_gray

    return img_conv
```

```python
path = "img1.jpg"

img = plt.imread(path)
grey = rgb2gray(img)
kernel = np.array([
    [0.0625, 0.125, 0.0625],
    [0.125, 0.1, 0.125],
    [0.0625, 0.125, 0.0625]
])

plt.title('AVANT')
plt.axis('off')
plt.imshow(grey, cmap='gray')
plt.show()

img_grey = convol(grey, kernel)

# imsave('blur_' + path, img_grey)

plt.title('APRÈS')
plt.axis('off')
plt.imshow(img_grey, cmap='gray')
plt.show()
```
